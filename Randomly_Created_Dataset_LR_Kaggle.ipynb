```python
#Importing the library 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Assigning the dataset into the variable 
test_ds = pd.read_csv("test.CSV")
print (test_ds)
train_ds = pd.read_csv("train.CSV")
print(train_ds)
```

```python
# Splitting the data set into dependent and independent variable 
X_test_ds = test_ds.iloc[:,:-1].values # Rows , Column
print(X_test_ds)
y_test_ds = test_ds.iloc[:,1:].values 
print(y_test_ds)
```

```python
# Splitting the data set into dependent and independent variable 
X_train_ds = train_ds.iloc[:,:-1].values # Rows , Column
print(X_train_ds)
y_train_ds = train_ds.iloc[:,1:].values 
print(y_train_ds)
```

```python
#Importing the library for Linear Regression 
from sklearn.linear_model import LinearRegression
simplelinearRegression=LinearRegression()
simplelinearRegression.fit(X_train_ds,y_train_ds)
```

```python
# Predicting the value
y_predict = simplelinearRegression.predict(X_test_ds)
```

```python
print(y_predict)
```

```python
#implement the graph
plt.scatter(X_train_ds,y_train_ds,color= 'green')
plt.plot(X_train_ds,simplelinearRegression.predict(X_train_ds))
plt.show()
```

```python
#implement the graph
plt.scatter(X_test_ds,y_test_ds,color= 'red')
plt.plot(X_test_ds,simplelinearRegression.predict(X_test_ds))
plt.show()
```

```python
y_predict_val_94  = simplelinearRegression.predict([[94.0]])
print(y_predict_val_94)
```
